## 3.5.5 Error Sources and Remaining Bias

### Residual OCR Errors

Despite filtering, OCR artifacts remain in the corpus:
- Misrecognized characters that resemble valid tokens (e.g., "rn" → "m", "0" → "O")
- Degraded page margins and low-contrast text introduce character-level noise
- Historical printing variations and font changes confuse modern OCR models
- Expected CER of 3-5% for cleanly printed pages, up to 10% for degraded originals

**Impact:** Token frequencies and rare word analysis are affected; common words unaffected

**Mitigation:** Applied lemmatization and stopword filtering to reduce noise propagation

---

### Bias Toward Dominant Fonts and Layouts

The OCR model was trained on modern, standardized documents:
- Body text (standard fonts) recognized with >95% accuracy
- Headers and sidebars recognized with <90% accuracy
- Italics and decorative fonts may be misrecognized
- Column layouts and multi-column text introduce confusion

**Impact:** Metadata extraction (headers, captions) less reliable than body text

**Mitigation:** Manual region segmentation in XML preprocessing; visual inspection of edge cases

---

### Marginal Text Complexity

Page margins, footers, and footnotes present specific challenges:
- Inconsistent spacing and alignment
- Small fonts and degraded quality
- Mixed content (text, page numbers, publisher marks)
- These are often filtered out by layout analysis, potentially losing content

**Impact:** Some marginal annotations and asides may be lost

**Mitigation:** Manual review of sample margins; documented in corpus preprocessing notes

---

### Under-recognition of Proper Names

Named entity recognition is imperfect for proper names:
- Capitalized common words misidentified as names (e.g., "Black Women" in artwork titles)
- Names with diacritics or non-English characters may be split or corrupted
- Abbreviated author names (initials only) are filtered by design
- Historical names unfamiliar to the model trained on contemporary data

**Impact:** Some theorists and artists may be under-counted; some false positives in captions

**Mitigation:** Manual review of top entities; preservation of raw variants for verification

---

### Structural Region Ambiguity

Separating reference lists from body text is not always clear-cut:
- Some bibliographies embedded within running text
- Mixed inline and formal citations
- Captions integrated with body text vs. visually separated

**Impact:** Some references may be counted as paragraph mentions and vice versa

**Mitigation:** XML preprocessing attempted disambiguation; some ambiguous cases remain

---

### Training Data Bias

The spaCy NER model is trained on contemporary English text:
- Limited exposure to 1970s-1990s academic and artistic discourse
- May miss or misidentify names of non-Western or marginalized figures
- Feminist theory terminology may not align with general English patterns
- Artist names and international contributors may be under-recognized

**Impact:** Potential bias toward recognizing mainstream, Anglo-American names

**Mitigation:** Manual normalization captures variants; comparison with external sources recommended

### Summary of Quality Tradeoffs

| Aspect | Strength | Limitation | Mitigation |
|--------|----------|-----------|------------|
| **OCR Accuracy** | 95%+ for clear text | 3-10% CER in margins/headers | Manual spot-checking |
| **Region Extraction** | Structured XML; 100% separation attempted | Ambiguous boundaries remain | Visual inspection of samples |
| **NER Performance** | Reliable for common names | Misses non-English, marginal text | Manual review + external verification |
| **Normalization** | Transparent lookup tables | Subjective decisions | Documented in thesis appendix |

### Remaining Bias Toward:
- Clear, well-printed text over degraded material
- Dominant English-language names over internationalized names
- Body text over marginal or decorative text
- Standard academic citation formats over informal references

These biases should be considered when interpreting temporal trends and attribution patterns.
