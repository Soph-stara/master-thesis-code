# Notebook 8 - Normalization

Normalization will only be applied to H1, H2 and H3 will be skipped.
For H1, normalization is safe because keyword analysis and not NER is being done.
For H2 and H3, NER is needed. Hence, I will need this without text normalization. 
Christine P. Chai explicitly warns lowercasing hurts NER.(Chai, 2023, p. 519)

### Sources
- Christine P. Chai. Comparison of text preprocessing methods. Natural Language Engineering, 29(3):509–553, 2023. doi: 10.1017/S1351324922000213.
- Python String lower() Method. (n.d.). Retrieved May 28, 2026, from https://www.w3schools.com/python/ref_string_lower.asp


### 1. Importing

```python
import pandas as pd
import ast
```

### 2. Loading the dataset

```python
df_h1 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/06_h1_punctuation.csv")
# df_h2 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/06_h2_punctuation.csv")
# df_h3 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/06_h3_punctuation.csv")

df_h1["tokens"] = df_h1["tokens"].apply(ast.literal_eval)
# df_h2["tokens"] = df_h2["tokens"].apply(ast.literal_eval)
# df_h3["tokens"] = df_h3["tokens"].apply(ast.literal_eval)
```

### 3. Applying the Lowercase function

Getting Help from this [Website](https://www.w3schools.com/python/ref_string_lower.asp).

```python
def to_lowercase(tokens):
    return [token.lower() for token in tokens]
```

### 4. Apply for H1 and Skipping H2 and H3

```python
df_h1["tokens"] = df_h1["tokens"].apply(to_lowercase)
# df_h3["tokens"] = df_h3["tokens"].apply(to_lowercase)
```
