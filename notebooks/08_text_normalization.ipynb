# Only Applying to H1 and H3, skipping H2!
H1 — yes, lowercase is safe. You are doing concept/keyword analysis, not NER.

H2 — this is the tricky one. You need NER to find person names and book titles. 

Chai explicitly warns lowercasing hurts NER. So for H2 you should skip lowercasing for now and do it after NER in 09_NER.ipynb.

H3 — your data is author_creator and caption fields which are short name strings. 

You actually want to lowercase here to normalize names like LIPPARD and Lippard into the same token.

#### 1. Importing

```python
import pandas as pd
import ast
```

#### 2. Loading the datasets

```python
df_h1 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/06_h1_punctuation.csv")
df_h2 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/06_h2_punctuation.csv")
df_h3 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/06_h3_punctuation.csv")

df_h1["tokens"] = df_h1["tokens"].apply(ast.literal_eval)
df_h2["tokens"] = df_h2["tokens"].apply(ast.literal_eval)
df_h3["tokens"] = df_h3["tokens"].apply(ast.literal_eval)
```

#### 3. Applying the Lowercase function
Using this: https://www.w3schools.com/python/ref_string_lower.asp

```python
def to_lowercase(tokens):
    return [token.lower() for token in tokens]
```

#### 4. Apply for H1 and H3 and Skipping H2

```python
df_h1["tokens"] = df_h1["tokens"].apply(to_lowercase)
# df_h2 is skipped on purpose — NER needs original casing
df_h3["tokens"] = df_h3["tokens"].apply(to_lowercase)
```

#### 5. Check (did it work?)

```python
print(df_h1["tokens"].iloc[0])
print("---")
print(df_h3["tokens"].iloc[0])
```

#### 6. Saving

```python
df_h1.to_csv("/Users/sophiehamann/master-thesis-code/data/output_files/07_h1_normalized.csv", index=False)
df_h2.to_csv("/Users/sophiehamann/master-thesis-code/data/output_files/07_h2_normalized.csv", index=False)  # unchanged but saved with new number
df_h3.to_csv("/Users/sophiehamann/master-thesis-code/data/output_files/07_h3_normalized.csv", index=False)
```
