# Notebook 6 - Tokenization

According to Christine P. Chai in her paper "Comparison of text preprocessing methods", tokenization is more complex than it seems and the choices made at this step affect everything downstream. The keypoints which are important for this corpus are, that one should not split on white space, but use NLTK's tokenizer instead. Additionally, Chai warns that undocumented tokenization decisions hurt reproducibility, so the aim is to document choices clearly in the process. (Chai, 2023, p. 515ff.).

### 1. Imports
Looking at the [NLTK](https://www.nltk.org/api/nltk.tokenize.html) Tokenization Package.

```python
import pandas as pd
import nltk

nltk.download("punkt")
nltk.download("punkt_tab")
```

### 2. Loading the Cleaned Corpora

```python
df_h1 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/04_h1_clean.csv")
df_h2 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/04_h2_clean.csv")
df_h3 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/04_h3_clean.csv")
```

#### 3. Writing the tokenization function

Getting Support from [this](https://www.geeksforgeeks.org/nlp/nlp-how-tokenizing-text-sentence-words-works/).

```python
from nltk.tokenize import word_tokenize

def tokenize_text(text):
    
    # skip if empty
    if pd.isna(text):
        return []
    
    # split text into individual word tokens
    tokens = word_tokenize(text)
    
    return tokens
```

### 4. Applying to all three corpora

```python
df_h1["tokens"] = df_h1["text"].apply(tokenize_text)
df_h2["tokens"] = df_h2["text"].apply(tokenize_text)
df_h3["tokens"] = df_h3["text"].apply(tokenize_text)
```

### 5. Checking the results

```python
# looking at the first row to see if tokenization looks right
print(df_h1["tokens"].iloc[0])
print("---")
# count total tokens per hypothesis
print(f"H1 total tokens: {df_h1['tokens'].apply(len).sum()}")
print(f"H2 total tokens: {df_h2['tokens'].apply(len).sum()}")
print(f"H3 total tokens: {df_h3['tokens'].apply(len).sum()}")
```
