### What Chai says about tokenization

Chai says tokenization is more complex than it seems and the choices you make here affect everything downstream. Key points for your corpus:

Don't just split on white space — use NLTK's tokenizer

Your corpus is historical text so expect unusual punctuation and spelling

Document your choices clearly — Chai warns that undocumented tokenization decisions hurt reproducibility

```python
import pandas as pd
import nltk

# download these once — you only need to run this cell once
nltk.download("punkt")
nltk.download("punkt_tab")
```

#### 2. Loading my cleaned corpora

```python
df_h1 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/04_h1_clean.csv")
df_h2 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/04_h2_clean.csv")
df_h3 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/04_h3_clean.csv")
```

#### 3. Writing the tokenization function

https://www.geeksforgeeks.org/nlp/nlp-how-tokenizing-text-sentence-words-works/

```python
from nltk.tokenize import word_tokenize

def tokenize_text(text):
    
    # skip if empty
    if pd.isna(text):
        return []
    
    # split text into individual word tokens
    tokens = word_tokenize(text)
    
    return tokens
```

#### 4. Applying to all three corpora

```python
df_h1["tokens"] = df_h1["text"].apply(tokenize_text)
df_h2["tokens"] = df_h2["text"].apply(tokenize_text)
df_h3["tokens"] = df_h3["text"].apply(tokenize_text)
```

#### 5. Checking the results

```python
# look at the first row to see if tokenization looks right
print(df_h1["tokens"].iloc[0])
print("---")
# count total tokens per hypothesis
print(f"H1 total tokens: {df_h1['tokens'].apply(len).sum()}")
print(f"H2 total tokens: {df_h2['tokens'].apply(len).sum()}")
print(f"H3 total tokens: {df_h3['tokens'].apply(len).sum()}")
```
