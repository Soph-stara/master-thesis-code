# Remove punctuation tokens from your token lists, while being careful to keep apostrophes in contractions and negations.

Chai warns specifically:

Don't remove punctuation too early or you lose sentence boundaries
Keep negations like "don't" and "won't" — removing them reverses meaning
Punctuation removal is permanent — you can't get it back

Since you are doing concept and reference analysis, not sentiment analysis, removing punctuation is safe here. You just need to protect contractions.

#### 1. Imports

```python
import pandas as pd
import string

print(string.punctuation)  # shows you exactly what will be removed
```

#### 2. Loading tokenized corpora

```python
import ast

df_h1 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/05_h1_tokenized.csv")
df_h2 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/05_h2_tokenized.csv")
df_h3 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/05_h3_tokenized.csv")

# the tokens column was saved as a string — this converts it back to a list
df_h1["tokens"] = df_h1["tokens"].apply(ast.literal_eval)
df_h2["tokens"] = df_h2["tokens"].apply(ast.literal_eval)
df_h3["tokens"] = df_h3["tokens"].apply(ast.literal_eval)
```

Using this: https://www.geeksforgeeks.org/python/python-convert-a-string-representation-of-list-into-list/

#### 3. Write Punctuation removal function
Using this: https://www.w3schools.com/python/python_lists_comprehension.asp

```python
special_tokens = {"``", "''", "'s", "n't", "--", "...", "'re", "'ve", "'ll", "'d", "'m"}

def remove_punctuation(tokens):
    cleaned = []
    for token in tokens:
        if token not in string.punctuation and token not in special_tokens:
            cleaned.append(token)
    return cleaned
```

#### 4. How does the punctuation look like in my tokens before removing them?

```python
# look at what punctuation tokens exist in your corpus
sample_tokens = df_h1["tokens"].iloc[0]
punct_tokens = [t for t in sample_tokens if t in string.punctuation]
print("Punctuation tokens found:", punct_tokens)
```

#### 5. Appyling to all three datasets

```python
df_h1["tokens"] = df_h1["tokens"].apply(remove_punctuation)
df_h2["tokens"] = df_h2["tokens"].apply(remove_punctuation)
df_h3["tokens"] = df_h3["tokens"].apply(remove_punctuation)
```

#### 6. Checking the results

```python
# compare token count before and after
print(f"H1 first row tokens: {df_h1['tokens'].iloc[0]}")
print(f"H1 total tokens: {df_h1['tokens'].apply(len).sum()}")
```
