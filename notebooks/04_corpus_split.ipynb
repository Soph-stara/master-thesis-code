# Splitting the Corpus by Hypothesis

H1/  H2/  H3/ ← separate folders, each with notebooks 05–11

Load your parsed corpus and split it into three separate CSV files based on which region types each hypothesis needs.

#### 1. Imports

```python
import pandas as pd
```

#### 2. Loading Parsed Corpus

```python
df = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/01_parsed_corpus.csv")
print(df.shape)
print(df["region_type"].value_counts())
```

#### 3. Splitting by Hypothesis

```python
# H1 needs: paragraph and heading
df_h1 = df[df["region_type"].isin(["paragraph", "heading"])]

# H2 needs: paragraph and reference_list
df_h2 = df[df["region_type"].isin(["paragraph", "reference_list"])]

# H3 needs: author_creator and caption
df_h3 = df[df["region_type"].isin(["author_creator", "caption"])]
```

#### 4. Checking if the splits worked

```python
print(f"H1: {len(df_h1)} regions")
print(f"H2: {len(df_h2)} regions")
print(f"H3: {len(df_h3)} regions")
```

#### 5. Save to csv

```python
df_h1.to_csv("/Users/sophiehamann/master-thesis-code/data/output_files/03_h1_corpus.csv", index=False)
df_h2.to_csv("/Users/sophiehamann/master-thesis-code/data/output_files/03_h2_corpus.csv", index=False)
df_h3.to_csv("/Users/sophiehamann/master-thesis-code/data/output_files/03_h3_corpus.csv", index=False)
```
