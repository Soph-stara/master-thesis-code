# Splitting the Corpus by Hypothesis

H1/  H2/  H3/ ← separate folders, each with notebooks 05–11

In this notebook, I am loading my parsed corpus (01_parsed_corpus.csv) and splitting it into three separate csv viles based on which region types each Hypothesis needs. 

#### 1. Imports

```python
import pandas as pd
```

#### 2. Loading Parsed Corpus

```python
df = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/01_parsed_corpus.csv")
print(df.shape)
print(df["region_type"].value_counts())
```

## 3. Splitting by Hypothesis
#### Hypothesis 1: 
Quantitative analysis of terminology and key concepts in Heresies will reveal distinct temporal shifts corresponding to major phases in feminist theory, including essentialist feminism in the late 1970s, the rise of French and difference feminism in the 1980s, and increasingly deconstructive theoretical orientations by the early 1990s.

Needed: paragraph, heading

#### Hypothesis 2: 
References to feminist texts and thinkers in Heresies will show changing patterns of prominence over time, reflecting a move from foundational feminist works toward French feminist theory and later deconstruction-oriented influences.

Needed: paragraph, reference_list

#### Hypothesis 3: 
Shifts in feminist theoretical orientation within Heresies will coincide with changes in patterns of contribution, including the emergence of new contributors and changing thematic emphases associated with the magazine’s rotating collectives.

Needed: author_creator, caption

```python
# Hypothsies 1
df_h1 = df[df["region_type"].isin(["paragraph", "heading"])]

# Hypothesis 2
df_h2 = df[df["region_type"].isin(["paragraph", "reference_list"])]

# Hypothesis 3
df_h3 = df[df["region_type"].isin(["author_creator", "caption"])]
```

#### 4. Checking if the splits worked

```python
print(f"H1: {len(df_h1)} regions")
print(f"H2: {len(df_h2)} regions")
print(f"H3: {len(df_h3)} regions")
```

#### 5. Save to csv

```python
df_h1.to_csv("/Users/sophiehamann/master-thesis-code/data/processed/03_h1_corpus.csv", index=False)
df_h2.to_csv("/Users/sophiehamann/master-thesis-code/data/processed/03_h2_corpus.csv", index=False)
df_h3.to_csv("/Users/sophiehamann/master-thesis-code/data/processed/03_h3_corpus.csv", index=False)
```
