#### 1. Imports

```python
import pandas as pd
import re
```

#### 2. Loading all three Corpora

```python
df_h1 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/03_h1_corpus.csv")
df_h2 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/03_h2_corpus.csv")
df_h3 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/03_h3_corpus.csv")
```

#### 3. Writing the Cleaning Function

Now I need to define a cleaning function (using regex mostly) to normalize the raw OCR/XML text into clean and readable strings.
 
I used [this](https://www.w3schools.com/python/python_regex.asp#sub) to help me write the code.

```python
def clean_text(text):
    
    # skip if empty
    if pd.isna(text):
        return text
    
    # removing XML or HTML tags e.g. <p> or <lb/>
    text = re.sub(r"<[^>]+>", " ", text)
    
    # fixing hyphenation at line breaks e.g. "histo-\nry" → "history"
    text = re.sub(r"-\n", "", text)
    
    # removing leftover line breaks
    text = re.sub(r"\n", " ", text)
    
    # fixing encoding issues (e.g. "â€™" → "’") 
    # through forcing the text through UTF-8 encode/decode, to silently drop any characters that can't be represented in UTF-8 
    text = text.encode("utf-8", "ignore").decode("utf-8")
    
    # removing extra whitespace
    text = re.sub(r"\s+", " ", text).strip()
    
    return text
```

#### 4. Applying this Function to all three Files

I looked at [this](https://pandas.pydata.org/docs/reference/api/pandas.Series.apply.html) for help.

```python
df_h1["text"] = df_h1["text"].apply(clean_text)
df_h2["text"] = df_h2["text"].apply(clean_text)
df_h3["text"] = df_h3["text"].apply(clean_text)
```
