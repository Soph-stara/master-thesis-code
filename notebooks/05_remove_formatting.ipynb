# Load each hypothesis CSV, clean the text, and save a new cleaned version.

#### 1. Imports
https://www.w3schools.com/python/python_regex.asp

```python
import pandas as pd
import re
```

#### 2. Loading all three Corpora

```python
df_h1 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/03_h1_corpus.csv")
df_h2 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/03_h2_corpus.csv")
df_h3 = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/output_files/03_h3_corpus.csv")
```

#### 3. Writing the Cleaning Function
https://www.w3schools.com/python/python_regex.asp#sub

```python
def clean_text(text):
    
    # skip if empty
    if pd.isna(text):
        return text
    
    # 1. remove any leftover XML or HTML tags e.g. <p> or <lb/>
    text = re.sub(r"<[^>]+>", " ", text)
    
    # 2. fix hyphenation at line breaks e.g. "histo-\nry" → "history"
    text = re.sub(r"-\n", "", text)
    
    # 3. remove leftover line breaks
    text = re.sub(r"\n", " ", text)
    
    # 4. remove non-UTF8 and garbled characters — keeps only basic readable characters
    text = text.encode("utf-8", "ignore").decode("utf-8")
    
    # 5. remove extra whitespace
    text = re.sub(r"\s+", " ", text).strip()
    
    return text
```

#### 4. Applying the Function to all three Files
https://pandas.pydata.org/docs/reference/api/pandas.Series.apply.html

```python
df_h1["text"] = df_h1["text"].apply(clean_text)
df_h2["text"] = df_h2["text"].apply(clean_text)
df_h3["text"] = df_h3["text"].apply(clean_text)
```

#### 5. Checking the Results

```python
# look at a few rows before and after to make sure it worked
print(df_h1["text"].iloc[0])
print("---")
print(df_h2["text"].iloc[0])
print("---")
print(df_h3["text"].iloc[0])
```

#### 6. Ssave cleaned versions

```python
df_h1.to_csv("/Users/sophiehamann/master-thesis-code/data/output_files/04_h1_clean.csv", index=False)
df_h2.to_csv("/Users/sophiehamann/master-thesis-code/data/output_files/04_h2_clean.csv", index=False)
df_h3.to_csv("/Users/sophiehamann/master-thesis-code/data/output_files/04_h3_clean.csv", index=False)
```
