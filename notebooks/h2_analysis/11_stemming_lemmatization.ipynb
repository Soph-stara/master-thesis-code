# Same as H1 — reduce words to their base form, but this time on the H2 token data.

#### 1. Imports
Using this: https://www.geeksforgeeks.org/python/python-lemmatization-with-nltk/

```python
import pandas as pd
import ast
import nltk

nltk.download("wordnet")
nltk.download("averaged_perceptron_tagger")
nltk.download("averaged_perceptron_tagger_eng")

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

#### 2. Loading Dataset

```python
df = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/10_h2_stopwords.csv")
df["tokens"] = df["tokens"].apply(ast.literal_eval)
print(df.shape)
```

#### 3. Helper and Lemmatizer function exaxtly as for H1

```python
lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def lemmatize_tokens(tokens):
    pos_tags = nltk.pos_tag(tokens)
    lemmatized = []
    for token, tag in pos_tags:
        wordnet_tag = get_wordnet_pos(tag)
        lemma = lemmatizer.lemmatize(token, wordnet_tag)
        lemmatized.append(lemma)
    return lemmatized
```

#### 4. Checking before and after on one row

```python
sample = df["tokens"].iloc[3]
print("BEFORE:", sample)
print("AFTER: ", lemmatize_tokens(sample))
```

#### 5. Applying the functions

```python
df["tokens"] = df["tokens"].apply(lemmatize_tokens)
print(f"Total tokens: {df['tokens'].apply(len).sum()}")
```

#### 6. Saving

```python
df.to_csv("/Users/sophiehamann/master-thesis-code/data/processed/11_h2_lemmatized.csv", index=False)
```
