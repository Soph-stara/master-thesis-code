# Notebook 9 H2 - Stopword Removal

Remove stopwords from the token column only, leaving the persons column completely untouched.

### 1. Imports

```python
import pandas as pd
import ast
import nltk
from nltk.corpus import stopwords

nltk.download("stopwords")
```

### 2. Load Datasets

```python
# load token data separately from person data
df_tokens  = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/06_h2_punctuation.csv")
df_persons = pd.read_csv("/Users/sophiehamann/master-thesis-code/data/processed/07_h2_persons.csv")

df_tokens["tokens"] = df_tokens["tokens"].apply(ast.literal_eval)

print(f"Token data: {df_tokens.shape}")
print(f"Person data: {df_persons.shape}")
```

### 3. Stopwords list

```python
stop_words = set(stopwords.words("english"))

# following Chai — never remove negations
negations   = {"no", "not", "nor", "never", "neither", "nothing", "nowhere", "nobody"}
boilerplate = {"idea-oriented", "devote", "heresies", "journal", "publication"}

stop_words = stop_words - negations
stop_words = stop_words | boilerplate
```

### 4. Lowercasing

```python
def to_lowercase(tokens):
    return [token.lower() for token in tokens]

df_tokens["tokens"] = df_tokens["tokens"].apply(to_lowercase)
```

### 4. Applying the stopwords list

```python
def remove_stopwords(tokens):
    return [token for token in tokens if token not in stop_words]

df_tokens["tokens"] = df_tokens["tokens"].apply(remove_stopwords)

print(f"Tokens before: check previous notebook")
print(f"Tokens after:  {df_tokens['tokens'].apply(len).sum()}")
```

### 5. checking if it worked

```python
print(df_tokens["tokens"].iloc[0])
```

### 6. Saving

```python
# save token data with stopwords removed
# person data does not need to be saved again — it is already in 07_h2_persons.csv
df_tokens.to_csv("/Users/sophiehamann/master-thesis-code/data/processed/07_h2_stopwords.csv", index=False)
```
